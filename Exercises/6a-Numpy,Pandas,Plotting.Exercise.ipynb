# Python Exercises - Numpy, Pandas & Plotting

## Setup: import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Problem 1


Create a pandas dataframe that contains the following columns:

Column Name | Values
--- | ---
A | 500 random numbers drawn from a normal distribution with mean 15 and standard deviation 5
B | 500 equally spaced numbers between 1 and 50 (inclusive)
C | 500 random integers drawn between 10 and 30 (inclusive)



```python
df = pd.DataFrame({
    "A" : np.random.normal(15,5,500),
    "B" : np.linspace(1,50,500),
    "C" : np.random.randint(10,30,500),

})
df
```

## Problem 2


Calculate the mean of column A.

```python
mean_A = df["A"].mean()
mean_A
```

## Problem 3


Plot a histogram of column A. It should contain a vertical line at the mean. Make the figure size (10,5). Make sure to add axis labels and a legend for the mean.




```python
n = int(500**(1/2))
plt.figure(figsize = (10,5))
plt.hist(df["A"],bins = n, edgecolor = "black")
plt.axvline(mean_A,color = "blue", label = "mean")
plt.xlabel("Column A")
plt.ylabel("Frequency")
plt.title("Title")
plt.legend()
plt.show()
```

## Problem 4


Create a line plot of the first 100 values in column C. Make sure to add axis labels.


```python
plt.plot(df["C"].iloc[:100])
plt.xlabel("Index")
plt.ylabel("Column C")
plt.show()
```

## Problem 5


Create a new column, `D`, that contains 3 times the values in column `B` plus some random noise (`np.random.normal(0,10,500`).

```python
df["D"] = 3*df["B"] + np.random.normal(0,10,500)
df
```

## Problem 6


Create a scatter plot with column `B` on the x-axis and column `D` on the y-axis.  Make sure to add axis labels.


```python
plt.scatter(df["B"], df["D"])
plt.xlabel("B column")
plt.ylabel("D column")
plt.show()
```
